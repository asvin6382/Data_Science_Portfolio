#### Functions

```python
def remove_pun(st):    
    
            word_tokens = word_tokenize(st)
            filtered_sentence_ = []
 
            for w in word_tokens :
                        if w not in string.punctuation:
                                 filtered_sentence_.append(w)
            return filtered_sentence_
        
def Clean_Text(filtered_sentence_):    
                    Cleaaned_=[]
                    for w in filtered_sentence_ :
                                  if w not in filtered_stopcwords:
                                           Cleaaned_.append(w)
                    return Cleaaned_
def ps(Cleaaned_):
                        Positive_Score=0
                        for i in range(len(pwords)):
                                 for j in range(len(Cleaaned_)):
                                              if pwords[i]==Cleaaned_[j]:
                                                          Positive_Score+=1
                        return  Positive_Score
def ns(Cleaaned_) :
    
                    Negative_Score=0
                    for i in range(len(nwords)):
                                             for j in range(len(Cleaaned_)):
                                                            if nwords[i]==Cleaaned_[j]:
                                                                                  Negative_Score+=1
                    return  Negative_Score

def asl(st,Total_Words_before_cleaning):
             sentences= nltk.sent_tokenize(st)
             Average_Sentence_Length = Total_Words_before_cleaning /len(sentences)
             return Average_Sentence_Length
    
def  NOCW(Cleaaned_):
                    count=0
                    for myword in Cleaaned_:
                                    d = {}.fromkeys('aeiou',0)
                                    haslotsvowels = False
                                    for x in myword.lower():
                                                   if x in d:
                                                         d[x] += 1
                                    for q in d.values():
                                                    if q > 2:
                                                         haslotsvowels = True
                                    if haslotsvowels:
                                                    count += 1
        
                    return count
    


def stop_words_nltk(filtered_sentence_):
                        tokens = []
 
                        for w in filtered_sentence_ :
                                          if w not in stop_words:
                                                        tokens.append(w)
                        return len(tokens)
def scpw(Cleaaned_):
                    stemmer = PorterStemmer()
                    words_=[]
                    for word in Cleaaned_:
                                  words_.append(stemmer.stem(word))
                    count_=0

                    for i in words_:
                                  for j in i.lower():
      
                                              if j =="a" or j=="e" or j=="i" or j== "o" or j=="u":
                                                                                      count_+=1
                    return count_
                
def pp(Cleaaned_):
                s=" "
                for i in Cleaaned_:
                                    s+=i

                pronounRegex = re.compile(r'I|we|my|ours|us',re.I)
                pronouns = pronounRegex.findall(s)
                count=0
                for i in pronouns:
                          if i =="US":
                                  count+=1

                return len(pronouns)-count
            
            
def AVL(l):
                    num=0
                    for i in range (len(l)):
                                  num+=len(l[i])
                    return num/len(l)
```

```python

#Cleaned Sentence by custoime stopwords:
for i in range(114):
    text=df.loc[i,"Text"]
    list_RP=remove_pun(text)
    list_CSW=Clean_Text(list_RP,)
    POSITIVE_SCORE=ps(list_CSW)
    NEGATIVE_SCORE=ns(list_CSW)
    POLARITY_SCORE=( POSITIVE_SCORE - NEGATIVE_SCORE)/ ((POSITIVE_SCORE + NEGATIVE_SCORE) + 0.000001)
    SUBJECTIVITY_SCORE=(( POSITIVE_SCORE - NEGATIVE_SCORE)/ (len(list_CSW)) + 0.000001)
    
    
    AVG_SENTENCE_LENGTH=asl(text,len(list_RP))
   
    PERCENTAGE_OF_COMPLEXWORDS=((NOCW(list_CSW))/(len(list_CSW)))
    
    FOG_INDEX=0.4 * (AVG_SENTENCE_LENGTH + PERCENTAGE_OF_COMPLEXWORDS)
    AVG_NUMBER_OF_WORDS_PER_SENTENCE=AVG_SENTENCE_LENGTH
    COMPLEX_WORD_COUNT=NOCW(list_CSW)
    WORD_COUNT=stop_words_nltk(list_CSW)
    SYLLABLE_PER_WORD=scpw(list_CSW)
    PERSONAL_PRONOUNS=pp(list_CSW)
    AVG_WORD_LENGTH=AVL(list_CSW)
    df.loc[i,"POSITIVE SCORE"]=POSITIVE_SCORE
    df.loc[i,"NEGATIVE SCORE"]= NEGATIVE_SCORE
    df.loc[i,"POLARITY SCORE"]=POLARITY_SCORE
    df.loc[i,"SUBJECTIVITY SCORE"]=SUBJECTIVITY_SCORE
    df.loc[i,"AVG SENTENCE LENGTH"]=int(AVG_SENTENCE_LENGTH)
    df.loc[i,"PERCENTAGE OF COMPLEX WORDS"]=PERCENTAGE_OF_COMPLEXWORDS
    df.loc[i,"FOG INDEX"]=FOG_INDEX
    df.loc[i,"AVG NUMBER OF WORDS PER SENTENCE"]=int(AVG_NUMBER_OF_WORDS_PER_SENTENCE)
    df.loc[i,"COMPLEX WORD COUNT"]=COMPLEX_WORD_COUNT
    df.loc[i,"WORD COUNT"]=WORD_COUNT
    df.loc[i,"SYLLABLE PER WORD"]=SYLLABLE_PER_WORD
    df.loc[i,"PERSONAL PRONOUNS"]=PERSONAL_PRONOUNS
    df.loc[i,"AVG WORD LENGTH"]=int(AVG_WORD_LENGTH)

df
```
